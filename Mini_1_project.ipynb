```python
# 1.Visualisation - SEABORN
import seaborn as sns
sns.distplot(df['depth']) 
```

```python
import matplotlib.pyplot as plt
f=plt.figure()
f.set_figwidth(35)
f.set_figheight(10)
plt.plot()
df['depth'].value_counts().plot(kind = 'bar')
plt.xticks(rotation = 'vertical')
plt.show()

```

```python
#to  count which are more in number
df['depth'].value_counts()
```

```python
import matplotlib.pyplot as plt
f=plt.figure()
f.set_figwidth(35)
f.set_figheight(10)
plt.plot()
sns.countplot(x = 'depth',data = df)
plt.xticks(rotation = 'vertical')
plt.show()

```

```python
#to find which are more in price
import matplotlib.pyplot as plt
f=plt.figure()
f.set_figwidth(35)
f.set_figheight(10)
plt.plot()
sns.barplot(x = df['depth'],y = df['price'])
plt.xticks(rotation = 'vertical')
plt.show()

```

```python
import numpy as np
np.min(df['depth'])
```

```python
np.max(df['depth'])
```
